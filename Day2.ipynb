```python
with open("input2.txt") as f:
    input = f.read().splitlines()


values = {
    "A":1,
    "B":2,
    "C":3,
    "X":1,
    "Y":2,
    "Z":3
}

score = 0
for game in input:
    # draw
    if(values[game[0]] == values[game[2]]):
        score+=3
    # win
    elif game == "A Y" or game == "B Z" or game == "C X":
        score+=6
    # add score
    score+=values[game[2]]

print(f"first strategy: {score}")

score = 0
for game in input:
    # draw
    if game[2] == "Y":
        score += values[game[0]] + 3
    # win
    elif game[2] == "Z":
        if game[0] == "C":
            score += 6+1
        else:
            score += 6+values[game[0]]+1
    # loose
    else:
        if game[0] == "A":
            score += 3
        else:
            score += values[game[0]]-1

print(f"second strategy {score}")
```

```python
import pandas as pd

def get_user_choice(player_1):
    
    user_choice = {
        "A": "rock",
        "B": "paper",
        "C": "scissors"
    }
    return user_choice.get(player_1)

def get_elf_choice(player_2):
    
    user_choice = {
        "X": "rock",
        "Y": "paper",
        "Z": "scissors"
    }
    return user_choice.get(player_2)
    

def calculate_score(player_1, player_2):
    score = 0
    
    user_choice = get_user_choice(player_1)
    elf_choice = get_elf_choice(player_2)
    
    if elf_choice == "rock":
        score += 1
    elif elf_choice == "paper":
        score += 2
    elif elf_choice == "scissors":
        score += 3
        
    if user_choice==elf_choice:
        print("Draw!")
        score += 3
    elif (user_choice == "rock" and elf_choice == "scissors") | (user_choice == "paper" and elf_choice == "scissors") | (user_choice == "scissors" and elf_choice == "rock"):
        print("You win!")
        score += 6
    else:
        print("You lose!")
        
    return score

# colnames = ["player_1","player_2"]
rock_paper_scissors = (
    pd.read_csv("input2.txt", header=None, names=["players"])
    .players.str.split(" ", expand=True)
    .rename({0: "player_1", 1: "player_2"}, axis=1)
    # .assign(
    #     score = lambda x: calculate_score(x["player_1"], x["player_2"])
    # )
)
```

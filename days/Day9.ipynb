```python
from aoc.utils import read_split_data
from collections import Counter
import numpy as np

data = read_split_data("data/input9.txt")
```

```python
class Part:
    
    def __init__(self, x: int, y: int, positions: list = None):
        self.x = x
        self.y = y
        self.positions = positions or list()

    def update_positions(self):
        self.positions.append((self.x, self.y))
            
class Head(Part):

    def move_right(self, steps: int):
        for i in range(1, steps+1):
            self.x = self.x + 1
            self.update_positions()
        
    def move_left(self, steps: int):
        for i in range(1, steps+1):
            self.x = self.x - 1
            self.update_positions()
            
    def move_down(self, steps: int):
        for i in range(1, steps+1):
            self.y = self.y - 1
            self.update_positions()
        
    def move_up(self, steps: int):
        for i in range(1, steps+1):
            self.y = self.y + 1
            self.update_positions()
        
    def move(self, direction: str, steps: int):
        self.direction = direction         
        self.steps = steps       
        
        if direction == "R":
            self.move_right(steps)
        elif direction == "L":
            self.move_left(steps)
        elif direction == "U":
            self.move_up(steps)
        elif direction == "D":
            self.move_down(steps)
        else:
            raise ValueError(f"Unknown direction {self.direction}")
        
        
class Tail(Part):
    
    def add_head_positions(self, head_positions):
        # There is only one head position to be added
        if len(head_positions) == 0:
            print("There are no head positions.")
        if type(head_positions) == tuple:
            self.positions += [head_positions]
        else:
            self.positions.extend(head_positions)
        
    def follow_head(self, head):
        
        # print(f"All head positions {head.positions}")
        head_positions = head.positions[-head.steps-1:-1] # The last steps + original head position
        print(f"Head positions: {head_positions}")
        
        if len(head_positions)>0:
            
            # Add all the last head positions except the current position of the tail
            add_not_all = (((head.direction == "L") or (head.direction=="R")) and (self.x==head.x)) or (((head.direction == "D") or (head.direction == "U")) and (self.y==head.y))

            if not add_not_all:
                head_positions_new = head_positions[1:]
                print(f"Go {head.direction}: {head_positions_new}")
                self.add_head_positions(head_positions_new)

            # Add all the head positions except but skip the first step
            elif add_not_all:
                if (self.x, self.y) in head_positions:
                    head_positions.remove((self.x, self.y))
                    print(f"New head positions: {head_positions}")
                    self.add_head_positions(head_positions)
                        
            self.x, self.y = self.positions[-1]
```

```python
# If it goes right or left the tail will be one step behind
    # Except if the head reaches the end of the grid
# If the head goes up once the tail will be at the same position
    # Otherwise the tail will go diagional
    
starting_position = (0, 0)
x, y = starting_position

head = Head(x, y)
tail = Tail(x, y)

print(f"(x,y): ({head.x},{head.y})")

for row in data:
    direction, steps = row.split(" ")
    print(direction, steps)
    
    print(f"Head (x,y): ({head.x},{head.y})")
    print(f"Tail (x,y): ({tail.x},{tail.y})")
    
    head.move(direction, int(steps))
    tail.follow_head(head)
    print(f"Direction: {head.direction} Steps: {head.steps} -> Head (x,y): ({head.x},{head.y}) Tail (x,y): ({tail.x},{tail.y})\n")
```

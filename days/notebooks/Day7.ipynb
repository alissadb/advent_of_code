```python
# Super clean: https://www.reddit.com/r/adventofcode/comments/zesk40/comment/iz9p30m/?context=3

directory_sizes: dict[str, int] = Counter()
directories = [""]

for line in data[1:]:

    if line == "$ cd ..":
        directories.pop()
    elif "$ cd " in line:
        directories.append(line.replace("$ cd ", ""))
    elif line.split(" ")[0].isnumeric():  # <value> <file_name>
        value, file_name = line.split(" ")

        for directory in Path("/".join(directories + [file_name])).parents:
            directory_sizes[str(directory)] += int(value)
```

```python
sum(size for size in directory_sizes.values() if size <= 100000)
```

```python
current_free_space = 70000000 - directory_sizes["\\"]
cleaned_space = 30000000 - current_free_space

next(size for size in sorted(directory_sizes.values()) if size >= cleaned_space)
```

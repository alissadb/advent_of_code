Day 1: Calorie Counting

```python
with open("input.txt") as f:
  elves = f.read()
  
calories_per_elf = [
  sum([
    int(calorie) 
    for calorie in elf.split("\n")
  ]) 
  for elf in elves.split("\n\n")
]
```

```python
calories_per_elf.sort(reverse=True)
print(f"Most calories per elf: {calories_per_elf[0]}")
print(f"Total calories top 3 Elves: {calories_per_elf[0]+calories_per_elf[1]+calories_per_elf[2]}")
```
